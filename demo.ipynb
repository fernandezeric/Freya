```python
import requests
import json
from astropy.io import ascii
from astropy.table import Table
```

```python
def consulta_degree(ra,dec,radius,format,catalog):
    url = 'http://0.0.0.0:5000/get_lc'
    """
    url = 'http://0.0.0.0:5000/get_lc_nearest'
    """
    args = {'ra':ra,'dec':dec,'radius':radius,'format':format,'catalogs':catalog}
    r = requests.get(url=url,params=args)
    return r
```

```python
rdg = consulta_degree(139.33444972,68.6350604,0.0002777,'csv','ps1')
```
